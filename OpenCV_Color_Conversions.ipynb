**Commonly used color conversion codes include:**

*   By Default it gives BGR Image
*   cv2.COLOR_BGR2RGB: Converts BGR (Blue, Green, Red) to RGB (Red,Green, Blue).
*   cv2.COLOR_BGR2GRAY: Converts BGR to grayscale.
*   cv2.COLOR_RGB2BGR: Converts RGB to BGR.



```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('28.jpg')
plt.imshow(img)
```

```python
# Read an image
img = cv2.imread('28.jpg')

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
```

```python
# Convert BGR to GRAY
img_rgb1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_rgb1)
```

```python
# Convert RGB to BGR
img_rgb2 = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

plt.imshow(img_rgb2)
```
